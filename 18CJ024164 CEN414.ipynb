```python
african_countries_me = ['Algeria',
                        'Angola',
                        'Benin',
                        'Botswana',
                        'Burkina Faso',
                        'Burundi',
                        'Cameroon',
                        'Cape Verde',
                        'Central African Republic',
                        'Chad',
                        'Comoros',
                        "Côte d'Ivoire",
                        'Democratic Republic of the Congo',
                        'Djibouti',
                        'Egypt',
                        'Equatorial Guinea',
                        'Eritrea',
                        'Eswatini',
                        'Ethiopia',
                        'Gabon',
                        'Gambia',
                        'Ghana',
                        'Guinea',
                        'Guinea-Bissau',
                        'Kenya',
                        'Lesotho',
                        'Liberia',
                        'Libyan Arab Jamahiriya',
                        'Madagascar',
                        'Malawi',
                        'Mali',
                        'Mauritania',
                        'Mauritius',
                        'Morocco',
                        'Mozambique',
                        'Namibia',
                        'Niger',
                        'Nigeria',
                        'Rwanda',
                        'Sao Tome and Principe',
                        'Senegal',
                        'Seychelles',
                        'Sierra Leone',
                        'Somalia',
                        'South Africa',
                        'South Sudan',
                        'Sudan',
                        'Tanzania',
                        'Togo',
                        'Tunisia',
                        'Uganda'
                        'Zambia',
                        'Zimbabwe'

                        ]


data = pandas.read_csv("C:\\un.csv")

data = data[data['Country Name'].isin(african_countries_me)]

data.set_index('Country Name', inplace=True)
data = data.sort_values('2009')
data
```

```python
for year in range(2009,2021):
    fig,ax = plt.subplots(1)
    data.plot.bar(ax = ax, y=str(year),figsize=(15,12), color='g', title=f"A GRAPH OF FIXED TELEPHONE SUBSCRIPTION OF COUNTRIES AGAINST THE COUNTRIES {year} ")
    
    ax.set_xlabel("COUNTRIES")
    ax.set_ylabel("NUMBER OF FIXED TELEPHONE SUBSCRIPTIONS")
    
```

```python
col = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2018','2019','2020']

```

```python
plt.bar(col,data[col].mean())
plt.xlabel('years')
plt.ylabel('mean')
plt.title('MEAN OF FIXED TELEPHONE SUBSCRIPTION')
plt.savefig('output2.png', dpi=300, bbox_inches='tight')
```

```python
plt.bar(col,data[col].median())
plt.xlabel('years')
plt.ylabel('median')
plt.title('MEDIAN OF FIXED TELEPHONE SUBSCRIPTION')
plt.savefig('output3.png', dpi=300, bbox_inches='tight')
```

```python
plt.bar(col,data[col].std())
plt.xlabel('years')
plt.ylabel('standard deviation')
plt.title('STANDATRD DEVIATION OF FIXED TELEPHONE SUBSCRIPTION')
plt.savefig('output4.png', dpi=300, bbox_inches='tight')
```

```python
plt.bar(col,data[col].var())
plt.xlabel('years')
plt.ylabel('Variance')
plt.title('VARIANCE OF FIXED TELEPHONE SUBSCRIPTION')
plt.savefig('output5.png', dpi=300, bbox_inches='tight')
```

```python
_range = data[col].max() - data[col].min()
_range.plot.line()
plt.ylabel("Range")
plt.xlabel("Years")
plt.title("RANGE OF FIXED TELEPHONE SUBSCRIPTION")
plt.savefig('output6.png', dpi=300, bbox_inches='tight')
```

```python
iqrs =[]
for year in range(2009,2021):
    iqrs.append(data[str(year)].quantile(0.75) - data[str(year)].quantile(0.25))
    
    
    
plt.ylabel("Interquatile range")
plt.xlabel("Years")
plt.title("INTERQUATILE RANGE FIXED SUBSCRIPTION")
plt.plot(range(2009,2021),iqrs)
plt.savefig('output7.png', dpi=300, bbox_inches='tight')
```
